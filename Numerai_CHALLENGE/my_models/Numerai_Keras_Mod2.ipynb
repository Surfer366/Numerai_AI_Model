# Preparing the data

```python
import pandas as pd

from matplotlib import pyplot as plt

```

```python
#CONSTANTS

ROUND = '238'
TOURNAMENT_NAME = "kazutsugi"
TARGET_NAME = 'target' #f"target_{TOURNAMENT_NAME}"
PREDICTION_NAME = f"prediction_{TOURNAMENT_NAME}"

```

#LOAD_DATASETS

```python
# download the latest training dataset (takes around 30s)

training_data = pd.read_csv("https://numerai-public-datasets.s3-us-west-2.amazonaws.com/latest_numerai_training_data.csv.xz").set_index('id')
training_data.head()
```

```python

# download the latest tournament dataset (takes around 30s)

testing_data = pd.read_csv("https://numerai-public-datasets.s3-us-west-2.amazonaws.com/latest_numerai_tournament_data.csv.xz").set_index('id')
testing_data.head()

```

```python
#DEFINE FEATURES

feature_names = [
    f for f in training_data.columns if f.startswith('feature')
]
print(f"Loaded {len(feature_names)} features columns")

train_Features = training_data[feature_names]

train_Target = training_data[TARGET_NAME]

print(':::::: train_Features ::::::', '\n', train_Features)
      
print(':::::: train_Target ::::::', '\n', train_Target)
```

# Define a NN in keras

```python
#IMPORT THE RELEVANT MODULE, SUB-MODULES
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
#BUILD THE MODEL

def build_model(learning_rate, layer_size):
    """Function to Build a Keras Model"""
    model = keras.Sequential([
        layers.Dense(layer_size, activation='relu',
                input_shape=[len(feature_names)]),
        layers.Dense(layer_size, activation='relu', kernel_regularizer='l2'),
        layers.Dense(1, activation='sigmoid')
]) 
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

    model.compile(loss='mse', optimizer=optimizer, metrics=['mae','mse'])

    return model

#Later called by
regressor_model = build_model(0.05, 100)
```

# Train the NN and validate by monitoring training losses


```python
def train_model(model, feature, label, epochs, batch_size):
    """Build a function to train the keras model"""
    history = model.fit(x=feature,
                       y=label,
                       batch_size=batch_size,
                       epochs=epochs)
   
    epochs = history.epoch
    hist = pd.DataFrame(history.history)
    mse = hist['mse']
    
    return epochs, mse

    epochs,mse = train_model(regressor_model, 
                         train_Features, 
                         train_Target, 
                         epochs, 
                         batch_size)
        
    def plot_the_loss_curve(epochs, mse):
        """Plot a curve of loss vs. epoch."""

        plt.figure()
        plt.xlabel("Epoch")
        plt.ylabel("Mean Squared Error")

        plt.plot(epochs, mse, label="Loss")
        plt.legend()
        plt.ylim([mse.min() * 0.95, mse.max() * 1.03])
        plt.draw()

    plot_the_loss_curve(epochs, mse)
```

# Prediction

```python
testing_data[PREDICTION_NAME] = regressor_model.predict(testing_data[feature_names])

df = testing_data[PREDICTION_NAME]

df.columns = ["id", "prediction_kazutsugi"]

df.to_csv('Desktop\surfer366\4_Work_HUB\Myfile.csv', header=True)

print(df)
```
